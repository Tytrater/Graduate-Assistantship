```python
import pandas as pd
# pd.set_option('display.max_colwidth', None) # dosplay full dataframe in output
# pd.set_option('display.max_rows', None) # instead, save output to .txt file
import numpy as np
from datetime import datetime, timedelta



def set_job(data,day,str_start,str_end,job):
    start = datetime.strptime(str_start,'%H:%M')
    end = datetime.strptime(str_end,'%H:%M')
    interval = timedelta(minutes=15)
    current = start
    while current <= end:
        data.at[pd.to_datetime(current).time(),day] = job
        current += interval
    return data

days = ['monday','tuesday','wednesday','thursday','friday','saturday','sunday']
time = pd.date_range('00:00','23:00',freq='15T').time
df = pd.DataFrame(columns=days,index=time)

df = set_job(df,'monday','11:30','14:15','IS491')
df = set_job(df,'monday','16:00','17:15','IS471')
df = set_job(df,'monday','19:00','21:45','MIS781')
df = set_job(df,'tuesday','17:30','18:45','ECO773')
df = set_job(df,'tuesday','19:00','21:00','ECO773')
df = set_job(df,'wednesday','11:30','14:15','IS491')
df = set_job(df,'wednesday','16:00','17:15','IS471')
df = set_job(df,'thursday','17:30','18:45','ECO773')
df = set_job(df,'thursday','19:00','21:00','ECO773')
df = set_job(df,'friday','11:30','14:15','MIS761')
df = set_job(df,'friday','15:15','16:30','IS491')
df = set_job(df,'friday','18:00','21:30','MIS761')
for day in days:
    set_job(df,day,'00:00','12:00','sleep')
    set_job(df,day,'12:15','12:45','meal')
    set_job(df,day,'13:00','14:15','swim')
    set_job(df,day,'14:30','15:00','meal')
    # set_job(df,day,'16:45','17:15','meal')
df.fillna('leisure',inplace=True)

goal_work = round(20/3,1) # 20 hours across 3 Courses
goal_course = 3*4 # 3 credits by 4 hours/credit
goal_leisure = 2*(5+4) # 2 hours/day for 5 workdays, and 4 hours/day for 2 workdays

hours = df.stack().value_counts()
hours = (hours*15/60) # convert from 15-min blocks to hours
hours_goals = [8*7,
               4*7,
               22,
               2*5,
               goal_work,goal_work,goal_work,
               goal_course,goal_course]
hours = pd.DataFrame(hours)
hours['goal'] = hours_goals
# hours['goal met?'] = hours > hours['goal'] # Not working
print(hours)
print(f'target hours: {hours["goal"].sum()}')
print(f'excess leisure: {51.50 - goal_leisure}\n')
df[-45:].style.apply(lambda x: ["background: red" if  v == 'leisure' else "" for v in x  ], axis = 1)
```
